## Sum the counts for different age categories

```python
# Compute the mortality categories for part4
morticd10_part4_cols_wo_nan_df['mort_1_4'] = morticd10_part4_cols_wo_nan_df[['Deaths3', 'Deaths4', 'Deaths5', 'Deaths6']].sum(axis=1)
morticd10_part4_cols_wo_nan_df['mort_5_24'] = morticd10_part4_cols_wo_nan_df[['Deaths7', 'Deaths8', 'Deaths9', 'Deaths10']].sum(axis=1)
morticd10_part4_cols_wo_nan_df['mort_25_44'] = morticd10_part4_cols_wo_nan_df[['Deaths11', 'Deaths12', 'Deaths13', 'Deaths14']].sum(axis=1)
morticd10_part4_cols_wo_nan_df['mort_45_64'] = morticd10_part4_cols_wo_nan_df[['Deaths15', 'Deaths16', 'Deaths17', 'Deaths18']].sum(axis=1)
morticd10_part4_cols_wo_nan_df['mort_65_84'] = morticd10_part4_cols_wo_nan_df[['Deaths19', 'Deaths20', 'Deaths21', 'Deaths22']].sum(axis=1)
morticd10_part4_cols_wo_nan_df['mort_85_95+'] = morticd10_part4_cols_wo_nan_df[['Deaths23', 'Deaths24', 'Deaths25']].sum(axis=1)

morticd10_part4_cols_wo_nan_df
```
